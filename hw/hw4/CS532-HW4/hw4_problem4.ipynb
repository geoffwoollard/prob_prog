```python
r_W_0 = np.zeros((T,L,np.array(r[0][0][0]).size))
r_b_0 = np.zeros((T,L,np.array(r[0][0][1]).size))
W_1_nrows, W_1_ncols = np.array(r[0][0][2]).shape
r_W_1 = np.zeros((T,L,W_1_nrows, W_1_ncols))
r_b_1 = np.zeros((T,L,np.array(r[0][0][3]).size))

for t in range(T):
    for l in range(L):
        r_W_0[t,l] = np.array(r[t][l][0]).flatten()
        r_b_0[t,l] = np.array(r[t][l][1]).flatten()
        r_W_1[t,l] = np.array(r[t][l][2])
        r_b_1[t,l] = np.array(r[t][l][3]).flatten()
```

```python
probs = np.exp(logW)
probs /= probs.sum()
```

```python
posterior_r_W_0 = (probs.reshape(T,L,1) * r_W_0).sum(axis=(0,1))
posterior_r_b_0 = (probs.reshape(T,L,1) * r_b_0).sum(axis=(0,1))
posterior_r_W_1 = (probs.reshape(T,L,1,1) * r_W_1).sum(axis=(0,1))
posterior_r_b_1 = (probs.reshape(T,L,1) * r_b_0).sum(axis=(0,1))


posterior_r2_W_0 = (probs.reshape(T,L,1) * r_W_0**2).sum(axis=(0,1))
posterior_r2_b_0 = (probs.reshape(T,L,1) * r_b_0**2).sum(axis=(0,1))
posterior_r2_W_1 = (probs.reshape(T,L,1,1) * r_W_1**2).sum(axis=(0,1))
posterior_r2_b_1 = (probs.reshape(T,L,1) * r_b_0**2).sum(axis=(0,1))

posterior_std_W_0 = np.sqrt(posterior_r2_W_0 - posterior_r_W_0**2)
posterior_std_b_0 = np.sqrt(posterior_r2_b_0 - posterior_r_b_0**2)
posterior_std_W_1 = np.sqrt(posterior_r2_W_1 - posterior_r_W_1**2)
posterior_std_b_1 = np.sqrt(posterior_r2_b_1 - posterior_r_b_1**2)
```

```python
sns.set_theme()

f, ax = plt.subplots(figsize=(12,1))
ax = sns.heatmap(posterior_r_W_0.reshape(1,-1),annot=posterior_std_W_0.reshape(1,-1))
plt.title('posterior $W1$ (expected $W1_{i}$ in color, std overlayed in digits)')
```

```python
f, ax = plt.subplots(figsize=(12,1))
ax = sns.heatmap(posterior_r_b_0.reshape(1,-1),annot=posterior_std_b_0.reshape(1,-1))
plt.title('posterior $b0$ (expected $b0_{i}$ in color, std overlayed in digits)')
```

```python
f, ax = plt.subplots(figsize=(12,12))
ax = sns.heatmap(posterior_r_W_1,annot=posterior_std_W_1)
plt.title('posterior $W1$ (expected $W1_{ij}$ in color, std overlayed in digits)')
```

```python
f, ax = plt.subplots(figsize=(12,1))
ax = sns.heatmap(posterior_r_b_1.reshape(1,-1),annot=posterior_std_b_1.reshape(1,-1))
plt.title('posterior $b1$ (expected $b1_{i}$ in color, std overlayed in digits)')
```

**Write a paragraph or two comparing and contrasting mean-field black-box variational inference to parameter estimation via gradient descent.**

Mean-field black-box VI desouples the parameters of each proposal distribution from the parameters in other distributions, in the gradients. This has the advantage that the gradients don't increase with the complexity of the proposals (their number and dependencies). On the other hand, as p. 132 of the course textbook [(van de Meent et al., 2021)](http://arxiv.org/abs/1809.10756) explains, the variance of the gradient estimator tends to be high, and so many samples per mini batch L and many iterations T are needed, more so if the model (i.e. program for our forward model) is more complex.

In contrast other methods like variational autoencoders (see Kingma and Welling, 2014; Rezende et al 2014 cited on p. 132 of the course textbook) can have low L, as low as L=1. 

An advantage of BBVI is that we don't need to know the analytical forms of the gradients w.r.t the ELBO (i.e. the gradient outside the expectation over the proposal). Instead, we only need to know the gradient of the distribution parameters being learned/optimized w.r.t. the respective proposal distribution at sampled data points. We know this because we specify the proposals. At the same time, if other methods that make restrictions on the proposals, can leverage the possibility of taking gradients w.r.t. the ELBO and avoid high variance of the gradients. See the [2018 review by Zhang et al](https://arxiv.org/pdf/1711.05597.pdf) for more discussion.
